# Data Cleaning:
Find the missing and inappropriate values, treat them appropriately. Remove all
duplicate rows. Find the outliers.
● Replace the value 0 in age as NaN
● Treat the null values in all columns using any measures(removing/ replace the
values with mean/median/mode)

```python
# Identify missing values
missing_values = df.isnull()
```

```python
# Count missing values
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
# Replace the value 0 in the "Employee_Age" column with NaN:
df['Employee_Age'].replace(0, np.nan, inplace=True)
```

```python

print("Numerical columns are:",df.select_dtypes(exclude=['object']).columns)
print("Categorical columns are:",df.select_dtypes(include=['object']).columns)
```

```python
# Treat the null values in all columns using the mean for numeric columns and the mode for categorical columns
numeric_columns = ['Employee_Age', 'Employee_Salary', 'Employee_Gender']
categorical_columns = ['Company_Name', 'Employee_Place', 'Employee_Country']

for column in numeric_columns:
    df[column].fillna(df[column].mean(), inplace=True)

for column in categorical_columns:
    df[column].fillna(df[column].mode().values[0], inplace=True)
```

```python
#Identifying duplicate rows
df[df.duplicated()]
```

```python
# Remove duplicate rows:
df.drop_duplicates(inplace=True)
```

```python
#checking for duplicate rows
print(df.shape)
# df[df.duplicated()]
```

```python
# Find outliers (using z-score or IQR method):
# Let's use the z-score method to identify outliers in the numeric columns (Employee_Age and Employee_Salary):
from scipy import stats

# Calculate the z-scores for numeric columns
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Define a threshold for outliers (e.g., z-score > 3 or z-score < -3)
threshold = 3

# Remove rows with outliers in any of the numeric columns
df_no_outliers = df[(z_scores < threshold).all(axis=1)]
df_no_outliers
```

# Data Analysis:
● Filter the data with age >40 and salary<5000
● Plot the chart with age and salary
● Count the number of people from each place and represent it visually


```python
# Filter the data based on age > 40 and salary < 5000:
filtered_data = df_no_outliers[(df_no_outliers['Employee_Age'] > 40) & (df_no_outliers['Employee_Salary'] < 5000)]
filtered_data
```

```python
plt.hist(filtered_data['Employee_Age'], bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age')

# Show the plot
plt.show()
```

```python
plt.hist(filtered_data['Employee_Salary'], bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary')

# Show the plot
plt.show()
```

```python
# Plot a scatter plot using age and salary columns:
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['Employee_Age'], filtered_data['Employee_Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs. Salary')
plt.show()
```

```python
# Count the number of people from each place and represent it visually:
place_counts = filtered_data['Employee_Place'].value_counts()
place_counts
```

```python
# Plotting the count of people from each place using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=place_counts.index, y=place_counts.values)
plt.xlabel('Place')
plt.ylabel('Count')
plt.title('Number of People from Each Place')
plt.xticks(rotation=45)
plt.show()
```

# Data Encoding:
Convert categorical variables into numerical representations using techniques such as
one-hot encoding, label encoding, making them suitable for analysis by machine
learning algorithms

```python
# One-hot encoding is suitable when the categorical variables have no ordinal relationship, meaning they don't have a natural order. 
# Using pandas get_dummies() function for one-hot encoding
encoded_data = pd.get_dummies(df_no_outliers, columns=['Company_Name', 'Employee_Place'], drop_first=True)
encoded_data
```

```python
# Label encoding is suitable when the categorical variables have an ordinal relationship, i.e., they have a meaningful order.
from sklearn.preprocessing import LabelEncoder

# Using LabelEncoder for label encoding
label_encoder = LabelEncoder()
df_no_outliers['Company_Name_LabelEncoded'] = label_encoder.fit_transform(df_no_outliers['Company_Name'])
df_no_outliers['Employee_Place_LabelEncoded'] = label_encoder.fit_transform(df_no_outliers['Employee_Place'])
```

# Feature Scaling:
After the process of encoding, perform the scaling of the features using standardscaler
and minmaxscaler.

```python
# Using StandardScaler:
from sklearn.preprocessing import StandardScaler

# Select the columns to be scaled (excluding the categorical variables)
numeric_columns = ['Employee_Age', 'Employee_Salary']

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the selected columns using StandardScaler
encoded_data[numeric_columns] = scaler.fit_transform(encoded_data[numeric_columns])
encoded_data[numeric_columns]
```

```python
# Using MinMaxScaler:
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
minmax_scaler = MinMaxScaler()

# Fit and transform the selected columns using MinMaxScaler
encoded_data[numeric_columns] = minmax_scaler.fit_transform(encoded_data[numeric_columns])
encoded_data[numeric_columns]
```
